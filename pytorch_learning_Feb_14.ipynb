붓꽃 데이터셋 

```python
import numpy as np
import matplotlib.pyplot as plt
import torch

# Define GELU function
def gelu(x):
    return x * 0.5 * (1 + torch.tanh(np.sqrt(2/np.pi) * (x + 0.044715 * x**3)))

# Define ReLU function
def relu(x):
    return torch.maximum(torch.tensor(0.0), x)

# Generate input values
x = torch.linspace(-5, 5, 200)

# Compute activations
gelu_y = gelu(x)
relu_y = relu(x)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, relu_y, label="ReLU", linestyle="dashed", linewidth=2, color="blue")
plt.plot(x, gelu_y, label="GELU", linewidth=2, color="red")
plt.axhline(0, color="black", linewidth=0.5, linestyle="dotted")
plt.axvline(0, color="black", linewidth=0.5, linestyle="dotted")
plt.legend()
plt.title("GELU vs ReLU Activation Functions")
plt.xlabel("Input (x)")
plt.ylabel("Activation Output")
plt.grid(True)
plt.show()
```
