```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
# Data
speed_kmh = [8, 9, 11.43, 12.54, 13, 14]
heart_rate_bpm = [120, 130, 138, 151, 156, 162]

# Fit a linear trendline
coeffs = np.polyfit(speed_kmh, heart_rate_bpm, 1)
trendline = np.poly1d(coeffs)

# Generate trendline points
speed_range = np.linspace(min(speed_kmh), max(speed_kmh), 100)
heart_rate_trend = trendline(speed_range)

# Plot with only points for data
plt.figure(figsize=(6, 4))
plt.scatter(speed_kmh, heart_rate_bpm, color='b', label='Heart Rate vs. Speed', zorder=2)
plt.plot(speed_range, heart_rate_trend, linestyle='--', color='r', label=f'Trendline (y = {coeffs[0]:.2f}x + {coeffs[1]:.2f})', zorder=1)

# Labels and title
plt.xlabel('Running Speed (km/h)')
plt.ylabel('Heart Rate (bpm)')
plt.title('Heart Rate vs. Running Speed with Trendline')
plt.grid(True)
plt.legend()
# Save as PNG and JPG
plt.savefig("heart_rate_vs_speed.png", dpi=300)  # Save as PNG
plt.savefig("heart_rate_vs_speed.jpg", dpi=300)  # Save as JPG

# Show plot
plt.show()
```
